## Exercise 4.5

### Simulate posterior model:

Work with N=100'000 simulations:

```r
N = 100000
```

Sample N values for $\pi$ from Beta distribution, then sample from binomial distribution with probability $\pi$ and $n=7$ and then restrict to outcomes where shop was still open at 2pm on exactly 3 days:

```r
simulate_posterior <- function( alpha, beta, n, y ) {
    data.frame( pi=rbeta(N, alpha, beta) ) %>% 
        mutate( days_open_at_2pm=rbinom(N, n, pi) ) %>% 
        filter( days_open_at_2pm==y )
}
```

For Kimya:

```r
posterior = simulate_posterior(1, 2, 7, 3)
hist(posterior$pi, breaks=100)
print(paste0(c("mean value = ", mean(posterior$pi))))
```

For Fernando:

```r
posterior = simulate_posterior(0.5, 1, 7, 3)
hist(posterior$pi, breaks=100)
print(paste0(c("mean value = ", mean(posterior$pi))))
```

For Ciara:

```r
posterior = simulate_posterior(3, 10, 7, 3)
hist(posterior$pi, breaks=100)
print(paste0(c("mean value = ", mean(posterior$pi))))
```

For Taylor:

```r
posterior = simulate_posterior(2, 0.1, 7, 3)
hist(posterior$pi, breaks=100)
print(paste0(c("mean value = ", mean(posterior$pi))))
```

## Exercise 4.13

### a)

Prior:

```r
pirange <- seq(0, 1, 0.001)
prior <- ifelse(pirange>=0.5, 2, 0)

options(repr.plot.width=10, repr.plot.height=5)
plot(pirange, prior, type="l")
```

### c)

Likelihood:

```r
likelihood <- (1-pirange)**100
plot( pirange, likelihood, type="l" )
```

Posterior:

```r
posterior <- prior*(1-pirange)**100

# make sure that posterior integrates to 1
posterior <- posterior / sum( posterior*0.001 )

plot( pirange, posterior, type="l" )
```
