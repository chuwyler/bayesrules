## Exercise 3.16

Posterior can easily be modeled with $\alpha = \frac{0.75}{0.25} \beta = 3 \beta$:

```r
k = 35
alpha2 = 3*k
beta2 = k
plot_beta( alpha=alpha2, beta=beta2 )
print( paste0(c("alpha2=", alpha2, ", beta2=", beta2 )) )
```

```r
alpha = 95
beta = 2
plot_beta(alpha=alpha, beta=beta)
```

```r
y = alpha2-alpha
n = beta2-beta+y
print( paste0(c("n=", n, ", y=", y )) )
plot_beta_binomial(alpha=alpha, beta=beta, n=n, y=y)
```

## Exercise 3.17

The prior is symmetric around $\pi=0.5$ and can be modelled with $\alpha=\beta$:

```r
k = 3
alpha = k
beta = k
plot_beta(alpha=alpha, beta=beta)
```

The posterior is centered around $\pi \sim 0.15\Rightarrow \alpha = \frac{0.15}{0.85} \beta \approx 0.18 \beta$:

```r
k = 80
alpha2 = round(0.18*k)
beta2 = k
plot_beta( alpha=alpha2, beta=beta2 )
print( paste0(c("alpha2=", alpha2, ", beta2=", beta2 )) )
```

```r
y = alpha2-alpha
n = beta2-beta+y
print( paste0(c("n=", n, ", y=", y )) )
plot_beta_binomial(alpha=alpha, beta=beta, n=n, y=y)
```
