## Exercise 5.11

```r
theta = 80
tau = 4
sigma = 3
```

### a)

```r
n1 = 32
ybar1 = 86
posterior_theta1 = theta * sigma^2 / (n1*tau^2 + sigma^2) + ybar1 * n1 * tau^2 / (n1*tau^2 + sigma^2)
posterior_tau1 = sqrt( tau^2 * sigma^2 / (n1*tau^2 + sigma^2) )
```

```r
posterior_theta1
```

```r
posterior_tau1
```

Verification:

```r
summarize_normal_normal(mean=theta, sd=tau, y_bar=ybar1, sigma=sigma, n=n1)
```

### b)

```r
n2 = 32
ybar2 = 82
posterior_theta2 = theta * sigma^2 / (n2*tau^2 + sigma^2) + ybar2 * n2 * tau^2 / (n2*tau^2 + sigma^2)
posterior_tau2 = sqrt( tau^2 * sigma^2 / (n2*tau^2 + sigma^2) )
```

```r
posterior_theta2
```

```r
posterior_tau2
```

Verification:

```r
summarize_normal_normal(mean=theta, sd=tau, y_bar=ybar2, sigma=sigma, n=n2)
```

### c)

#### Variant 1: Sequentally

```r
posterior_theta = posterior_theta1 * sigma^2 / (n2*posterior_tau1^2 + sigma^2) + ybar2 * n2 * posterior_tau1^2 / (n2*posterior_tau1^2 + sigma^2)
posterior_tau = sqrt( posterior_tau1^2 * sigma^2 / (n2*posterior_tau1^2 + sigma^2) )
```

```r
posterior_theta
```

```r
posterior_tau
```

#### Variant 2: By computing combined mean

```r
ybar = (n1 * ybar1 + n2 * ybar2) / (n1+n2)
ybar
```

```r
n = n1 + n2
n
```

```r
summarize_normal_normal(mean=theta, sd=tau, y_bar=ybar, sigma=sigma, n=n)
```
